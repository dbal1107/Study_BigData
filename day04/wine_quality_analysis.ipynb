### t-검정 / 회귀분석

#### 라이브러리 import

```python
from scipy import stats
```

```python
# statsmodels 패키지 추가
!pip install statsmodels
```

##### 회귀분석을 위한 statsmodels 패키지 import

```python
from statsmodels.formula.api import ols, glm
```

```python
red_wine_quality = wine.loc[wine.type=='red','quality']
red_wine_quality # wine
```

```python
white_wine_quality = wine.loc[wine['type']=='white','quality']
white_wine_quality
```

```python
# 3.5교시
# t-검정 (두 그룹 사이의 차이)
stats.ttest_ind(red_wine_quality, white_wine_quality, equal_var=False)
# 10 red-white 차이, pvalue :오차범위 (낮다)
```

```python
wine.columns
```

```python
# 선형회귀분석
Rformula = 'quality ~ fixed_acidity + volatile_acidity + citric_acid + residual_sugar + chlorides + free_sulfur_dioxide + total_sulfur_dioxide + density + pH + sulphates + alcohol'
```

```python
regre_result = glm(Rformula, data=wine).fit()
regre_result.summary()
```

```python
# OLS : 이상리스트스퀘어모듈
regre_result = ols(Rformula, data=wine).fit()
regre_result.summary()
```

### OLS 모델로 계산된 회귀분석 결과로 새 샘플(와인)에 대한 품질등급 예측

```python

sample1 = wine[wine.columns.difference(['quality', 'type'])]
sample1 = sample1[0:5][:]
sample1
```

```python
# quality 예측값
sample1_predict = regre_result.predict(sample1)
sample1_predict
```

### 5교시

#### 선형회귀 분석값 시각화

```python
# 패키지 import
import statsmodels.api as sm

others = list(set(wine.columns).difference(set(['quality','fixed_acidity'])))
others
```

```python
#고정산의 값에 따라서 품질에 영향을 미치는가
p, resids = sm.graphics.plot_partregress('quality','fixed_acidity', others, data=wine, ret_coords=True)
plt.show()
```

```python
(p, resids)
```
